The only important outliers for this data are going to be in the lyrics column since we verified that the other columns data are not neccessary for this dataset since the Spotify dataset contains the appropriate data. This boxplot shows us that the there are mainly outliers that have a higher than average word count. Most of the outliers can be written off as songs that simply have a lot of lyrics, like rap, or longer songs.
```python
#function to ensure that the lyrics are in english
def get_language(text):
    try:
        return detect(text)
    except:
        return 'idk'

#cleaning the data
#Removing duplicates
df_billboard_clean = df_billboard_tidy.drop_duplicates(subset=['year','ranking'], keep='first')
#ensures that the song isnt ridiculously long or too short
df_billboard_clean = df_billboard_clean[df_billboard_clean['word_count'] < 1200] #I determined that the songs with higher word count were a little messed up
df_billboard_clean = df_billboard_clean[df_billboard_clean['word_count'] > 15]
#Make sure everything is in english
df_billboard_clean['language'] = df_billboard_clean['lyrics'].apply(get_language)
df_billboard_clean = df_billboard_clean[df_billboard_clean['language'] == 'en']
rows = df_billboard_clean.shape[0]
columns = df_billboard_clean.shape[1]
print("The new number of rows is: ", rows)
print("The new number of columns is: ", columns)
print("The longest song is: ", df_billboard_clean['word_count'].max(), " words")
print("The shortest song is: ", df_billboard_clean['word_count'].min(), " words")
df_billboard_clean.head()
```

We stated why we previously did not remove null values, and after removing duplicate songs, unrealistically wordy songs, songs with few words, and songs that are not in English, our dataset is now clean and tidy.

Since our project primarily focuses 'escapism' using the sentiment of song lyrics, we will primarily foucs on data set #2. ### Unemployment rate
Once again taken from the Federal Reserve Economic Database, this data consists of a year by year record of the unemployment rate from 1948 to early 2026. #### Keyword-based Theme Analysis and Economic Correlations

To explore the relationship between popular music and the economic climate, we constructed three word lists which encapsulate escapist sentiment. Each song in the Billboard dataset was scored based on the frequency of theme-related keywords appearing in its lyrics.

```python
### We will be creating a word list of common party words, romance words, and escapist words
    # We need to figure out a better way to get these keywords as right now they were chosen at random from the internet
party_words = ['party', 'dance', 'club', 'drunk', 'night', 'celebrate', 'wild', 'smoke']
romance_words = ['love', 'kiss', 'heart', 'hold', 'forever', 'babe', 'mine', 'baby', 'darling', 'together']
escape_words = ['free', 'fly', 'high', 'lose', 'numb', 'forget', 'away', 'escape', 'dream']

# Count the number of times theme-related words appear in a song's lyrics
def keyword_score(lyrics, keywords):
    lyrics = str(lyrics).lower()
    return sum(lyrics.count(word) for word in keywords)

# Applying scoring to each song per theme
billboard_df['party_score'] = billboard_df['clean_lyrics'].apply(lambda x : keyword_score(x, party_words))
billboard_df['romance_score'] = billboard_df['clean_lyrics'].apply(lambda x : keyword_score(x, romance_words))
billboard_df['escape_score'] = billboard_df['clean_lyrics'].apply(lambda x : keyword_score(x, escape_words))
```

```python
# Aggregate keyword scores by year to observe shifts over time
yearly_df = billboard_df.groupby('year')[['party_score', 'romance_score', 'escape_score']].mean().reset_index()
yearly_df
```

The dataframe above shows how the prevalence of each theme has shifted over time through Billboard charting songs. We will now visualize these trends to better identify patterns over time.

```python
# Plot the average keyword scores for each theme over time
plt.figure(figsize=(12, 6))
plt.plot(yearly_df['year'], yearly_df['party_score'], label='Party')
plt.plot(yearly_df['year'], yearly_df['romance_score'], label='Romance')
plt.plot(yearly_df['year'], yearly_df['escape_score'], label='Escape')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.title('Song Theme Trends Over Time')
plt.legend()
plt.show()
```

The chart reveals that romance-themed language has been the dominant theme throughout the 2000-2023 period, though it shows a clear long-term decline. Party and escape themes remain relatively stable and low throughout the period, with a notable party spike around 2010-2011 coinciding with the post-financial crisis recovery.

```python
#Extract year from quarterly GDP data and aggregate to a new dataframe
gdp_df['year'] = pd.to_datetime(gdp_df['observation_date']).dt.year
gdp_yearly = gdp_df.groupby('year')['GDP'].mean().reset_index()
#Extract year from monthly unemployment data and aggregaete to a new dataframe
unemployment_df['year'] = pd.to_datetime(unemployment_df['observation_date']).dt.year
unemployment_yearly = unemployment_df.groupby('year')['UNRATE'].mean().reset_index()
```

```python
#Merge the Billboard, GDP, and Unemployment Yearly Dataframes
merged_df = yearly_df.merge(gdp_yearly, on = 'year').merge(unemployment_yearly, on = 'year')
merged_df['gdp_growth'] = merged_df['GDP'].pct_change() * 100
print(merged_df.head())
```

With our theme scores and economic indicators now aligned by year, we  merge them into a 
single dataframe and can now examine the relationships between musical themes and economic conditions. 

To examine these relationships, we will be using a correlation matrix to show the pairwise correlations between all variables, helping 
us identify which themes have the strongest relationships with GDP and unemployment.

```python
# Heatmap of correlations between theme scores and economic indicators
# Red indicates positive correlation, blue indicates negative correlation
corr_cols = ['party_score', 'romance_score', 'escape_score', 'gdp_growth', 'UNRATE']
plt.figure(figsize=(10, 8))
sns.heatmap(merged_df[corr_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

The correlation matrix reveals several relationships between musical themes and economic indicators. Party themes show the highest correlation with unemployment (r = 0.54), suggesting that during periods of high unemployment, popular music tends to feature more party-related language. Romance themes also show a positive correlation with unemployment (r = 0.40), meaning that during times of economic hardship, popular music tends to have an increase in romance-themed language. Escapist themes show a weaker correlation (r = 0.09), suggesting that escapist language is less tied to the economic conditions of the time.

Given these findings, we chose to plot each theme against unemployment as it showed the strongest relationships across all three themes. The following plots visualize this relationship over time.

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 12))

# Party Score vs Unemployment PLOT
axes[0].plot(merged_df['year'], merged_df['party_score'], label='Party', color='blue')
ax2 = axes[0].twinx()
ax2.plot(merged_df['year'], merged_df['UNRATE'], label='Unemployment', color='red', linestyle='--')
axes[0].set_title('Party Score vs Unemployment')
axes[0].set_ylabel('Party Score')
ax2.set_ylabel('Unemployment Rate')

# Romance Score vs Unemployment PLOT
axes[1].plot(merged_df['year'], merged_df['romance_score'], label='Romance', color='orange')
ax3 = axes[1].twinx()
ax3.plot(merged_df['year'], merged_df['UNRATE'], label='Unemployment', color='red', linestyle='--')
axes[1].set_title('Romance Score vs Unemployment')
axes[1].set_ylabel('Romance Score')
ax3.set_ylabel('Unemployment Rate')

# Escape Score vs Unemployment PLOT
axes[2].plot(merged_df['year'], merged_df['escape_score'], label='Escape', color='green')
ax4 = axes[2].twinx()
ax4.plot(merged_df['year'], merged_df['UNRATE'], label='Unemployment', color='red', linestyle='--')
axes[2].set_title('Escape Score vs Unemployment')
axes[2].set_ylabel('Escape Score')
ax4.set_ylabel('Unemployment Rate')

plt.tight_layout()
plt.show()
```

The plots above reveal some interesting visual patterns. Party score appear to be in line with 
unemployment, with both rising sharply around the 2008 financial crisis and peaking around 
2010-2011 before declining together. This supports the correlation finding (r=0.54).

Romance score shows a less consistent relationship with unemployment. While both spike 
around 2010, romance shows a long term decline that unemployment does not follow.

Escape score appears the most erratic, with no clear visual relationship 
with unemployment, consistent with its weak correlation (r=0.09). This suggests escapist 
language in popular music may be driven by factors other than economic conditions.
